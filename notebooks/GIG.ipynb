```python
import numpy as np
from scipy.stats import gennorm
import pylab as plt

%matplotlib widget

fig, ax = plt.subplots(1, 1, dpi=300)
beta = 1.3
mean, var, skew, kurt = gennorm.stats(beta, moments='mvsk')
x = np.linspace(gennorm.ppf(0.01, beta), gennorm.ppf(0.99, beta), 100)

ax.plot(x, gennorm.pdf(x, beta), 'r-', lw=5, alpha=0.6, label='Laplacian')

rv = gennorm(beta)
ax.plot(x, rv.pdf(x), 'k-', lw=2)
gennorm.rvs(beta, size=1000)
r = gennorm.rvs(beta, size=1000)
ax.hist(r, density=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)
plt.title(r'$\beta$={}'.format(beta))
plt.show()
```
